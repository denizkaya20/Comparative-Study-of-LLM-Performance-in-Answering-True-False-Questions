```python
import pandas as pd
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Dosyayı yükleme
file_path = '/content/sample_data/data_final.csv'
data = pd.read_csv(file_path)

# Gerçek cevapları ve modellerin tahminlerini çıkarma
true_labels = data['answer']
zephyr_predictions = data['Zephyr']
llama2_predictions = data['Llama2']
mistral_predictions = data['Mistral']

# Her bir model için metrikleri hesaplama
metrics = {
    'Model': ['Zephyr', 'Llama2', 'Mistral'],
    'Accuracy': [
        accuracy_score(true_labels, zephyr_predictions),
        accuracy_score(true_labels, llama2_predictions),
        accuracy_score(true_labels, mistral_predictions)
    ],
    'Precision': [
        precision_score(true_labels, zephyr_predictions),
        precision_score(true_labels, llama2_predictions),
        precision_score(true_labels, mistral_predictions)
    ],
    'Recall': [
        recall_score(true_labels, zephyr_predictions),
        recall_score(true_labels, llama2_predictions),
        recall_score(true_labels, mistral_predictions)
    ],
    'F1 Score': [
        f1_score(true_labels, zephyr_predictions),
        f1_score(true_labels, llama2_predictions),
        f1_score(true_labels, mistral_predictions)
    ]
}

# Metrikleri daha okunabilir hale getirmek için DataFrame'e çevirme
metrics_df = pd.DataFrame(metrics)

# Metrikleri görüntüleme
print(metrics_df)
```
