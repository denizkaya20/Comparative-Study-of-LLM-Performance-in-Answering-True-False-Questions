```python
# import dependencies
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer, BitsAndBytesConfig, pipeline

import os
from google.colab import drive

import chromadb
from langchain.llms import HuggingFacePipeline
from langchain.document_loaders import TextLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import Chroma
from langchain import HuggingFacePipeline
from langchain.document_loaders import PyPDFDirectoryLoader
from langchain.chains import ConversationalRetrievalChain
from langchain.chains import RetrievalQA
from langchain.memory import ConversationBufferMemory
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline
```

```python
# specify model huggingface mode name
model_name = "anakin87/zephyr-7b-alpha-sharded"

def load_quantized_model(model_name: str):
  model = AutoModelForCausalLM.from_pretrained(model_name, device_map="auto", load_in_8bit=True)
  return model
```

```python
# fucntion for initializing tokenizer
def initialize_tokenizer(model_name: str):
    """
    Initialize the tokenizer with the specified model_name.

    :param model_name: Name or path of the model for tokenizer initialization.
    :return: Initialized tokenizer.
    """
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    tokenizer.bos_token_id = 1  # Set beginning of sentence token id
    return tokenizer
```

```python
# load model
model = load_quantized_model(model_name)

# initialize tokenizer
tokenizer = initialize_tokenizer(model_name)

# specify stop token ids
stop_token_ids = [0]
```

```python
from langchain_community.document_loaders.csv_loader import CSVLoader
loader = CSVLoader(file_path="/content/sample_data/boolq_train.csv",source_column="passage")
docs_all = loader.load()
```

```python

# split the documents in small chunks
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=100) #Chage the chunk_size and chunk_overlap as needed
all_splits = text_splitter.split_documents(docs_all)

# specify embedding model (using huggingface sentence transformer)
embedding_model_name = "sentence-transformers/all-mpnet-base-v2"
model_kwargs = {"device": "cuda"}
embeddings = HuggingFaceEmbeddings(model_name=embedding_model_name, model_kwargs=model_kwargs)

#embed document chunks
vectordb = Chroma.from_documents(documents=all_splits, embedding=embeddings, persist_directory="chroma_db")

# specify the retriever
retriever = vectordb.as_retriever()
```

```python
from langchain.chains import RetrievalQAWithSourcesChain
```

```python
#pipe = pipeline("text-generation", model=model, tokenizer=tokenizer)
```

```python
pipe = pipeline("text-generation",
                model=model,
                tokenizer= tokenizer,
                torch_dtype=torch.bfloat16,
                device_map="auto",
                max_new_tokens = 1024,
                do_sample=True,
                top_k=10,
                num_return_sequences=1,
                eos_token_id=tokenizer.eos_token_id
                )
```

```python
# specify the llm
llm=HuggingFacePipeline(pipeline=pipe, model_kwargs={'temperature':0})
```

```python
chain = RetrievalQA.from_chain_type(llm=llm, chain_type='stuff', retriever=vectordb.as_retriever(search_kwargs={"k":6}), verbose=True,  return_source_documents=False,chain_type_kwargs={"verbose": True})
```

```python
import pandas as pd
file_path = '/content/sample_data/boolq_train.csv'
data = pd.read_csv(file_path)
data_son=data.head(500) #İlk 500 veriyi alıyoruz modele sokmak için
```

```python
from itertools import count
from langchain.output_parsers import CommaSeparatedListOutputParser
from langchain.prompts import PromptTemplate
from langchain.output_parsers.list import ListOutputParser

data = data_son
keywords = ["True", "False", "Yes", "No", "yes", "no", "true", "false"]
responses = []
explanation = []
count = 0

for index, row in data.iterrows():
    query = f"{row['question']}"
    result = chain(query)

    output_parser = CommaSeparatedListOutputParser()
    format_instructions = output_parser.get_format_instructions()

    output = output_parser.parse(result["result"])
    parsed_output = output

    # Tüm metni tek bir string olarak birleştir
    full_text = "\n".join(parsed_output)

    # '</s>\n' ifadesinden sonraki kısımda 'Option' ile başlayan cümleyi bul
    separator_index = full_text.find("\n\n")
    if separator_index != -1:
        post_separator_text = full_text[separator_index + len("\n\n"):]  # Ayırıcıdan sonraki metni al
        lines = post_separator_text.split("\n")  # Metni satırlara böl
        for line in lines:
            line = line.strip()  # Her satırı temizle
            if line.startswith("Helpful Answer:"):
                # İlk 'Option' ile başlayan satırı yazdır ve döngüyü sonlandır
                end_index = line.find(".")
                if end_index != -1:
                    #explanation_text = line[:end_index + 1]
                    explanation.append(result["result"])
                else:
                    explanation_text = line
                    #explanation.append(explanation_text)
                    explanation.append(result["result"])
                # Anahtar kelimeleri kontrol et ve responses'a ekle
                for keyword in keywords:
                    if keyword in explanation_text:
                        responses.append(keyword)
                        break  # Bulunduğu anda döngüden çık (birden fazla anahtar kelime olmasını istemiyorsak)
                break
    else:
        print("Aranan ayırıcı metinde bulunamadı.")

    count += 1
    print(count)

print("İşlem tamamlandı.")
```

```python
import csv

with open('/content/data.csv', 'r+', newline='') as csvfile:
    reader = csv.reader(csvfile)  # Okuyucu nesnesi
    all_rows = list(reader)  # Tüm satırları bir listeye dönüştürüyoruz

    if len(all_rows) > 0:
        all_rows[0].append('Zephyr_content')
        all_rows[0].append('Zephyr')

        # explanations ve responses listesindeki her elemanı ilgili satıra ekliyoruz
        for i in range(max(len(explanation), len(responses))):
            if i + 1 < len(all_rows):
                if i < len(explanation):
                    all_rows[i + 1].append(explanation[i])
                else:
                    all_rows[i + 1].append('')  # Eğer explanations listesi daha kısa ise, boş değer ekliyoruz

                if i < len(responses):
                    all_rows[i + 1].append(responses[i])
                else:
                    all_rows[i + 1].append('')  # Eğer responses listesi daha kısa ise, boş değer ekliyoruz
            else:
                # Eğer explanations veya responses listesi satır sayısından uzunsa, yeni satırlar ekliyoruz
                new_row = [''] * (len(all_rows[0]) - 2)
                if i < len(explanation):
                    new_row.append(explanation[i])
                else:
                    new_row.append('')

                if i < len(responses):
                    new_row.append(responses[i])
                else:
                    new_row.append('')

                all_rows.append(new_row)

        # Dosya imlecini dosyanın başına getiriyoruz
        csvfile.seek(0)

        # Yazar nesnesi oluşturuyoruz
        writer = csv.writer(csvfile)

        # Güncellenmiş satırları dosyaya yazıyoruz
        writer.writerows(all_rows)

        # CSV dosyasının kalan kısmını temizliyoruz (eğer varsa)
        csvfile.truncate()

print("CSV güncelleme tamamlandı.")
```
