```python
from llama_index.core import SimpleDirectoryReader, ServiceContext, StorageContext, VectorStoreIndex
from llama_index.llms.huggingface import HuggingFaceLLM
#from llama_index.embeddings.huggingface import HuggingFaceEmbedding
from llama_index.core import VectorStoreIndex
#from llama_index.vector_stores.chroma import ChromaVectorStore
import chromadb
from IPython.display import Markdown, display
from langchain import HuggingFacePipeline
from langchain.vectorstores import Chroma
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.chains import RetrievalQA
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.chains import RetrievalQAWithSourcesChain
from huggingface_hub import login
import torch
from transformers import BitsAndBytesConfig
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline
from langchain_community.document_loaders.csv_loader import CSVLoader
from langchain.embeddings import HuggingFaceEmbeddings
import pandas as pd
from itertools import count
from langchain.output_parsers import CommaSeparatedListOutputParser
from langchain.prompts import PromptTemplate
from langchain.output_parsers.list import ListOutputParser
```

```python
loader = CSVLoader(file_path="/content/sample_data/Llama2.csv",source_column="passage")
documents = loader.load()
```

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=100)
text_chunks = text_splitter.split_documents(documents)
```

```python
login("API KEY", add_to_git_credential=True)
```

```python
quantization_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_compute_dtype=torch.float16,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_use_double_quant=True,
)
```

```python
model_id = "mistralai/Mistral-7B-Instruct-v0.1"
```

```python
import torch
torch.set_default_device('cuda')
```

```python
model_4bit = AutoModelForCausalLM.from_pretrained( model_id, device_map="auto",quantization_config=quantization_config, )
tokenizer = AutoTokenizer.from_pretrained(model_id)
```

```python
pipeline = pipeline(
        "text-generation",
        model=model_4bit,
        tokenizer=tokenizer,
        use_cache=True,
        device_map="auto",
        max_length=2056,
        do_sample=True,
        top_k=5,
        num_return_sequences=1,
        eos_token_id=tokenizer.eos_token_id,
        pad_token_id=tokenizer.eos_token_id,
)
```

```python
import torch
torch.set_default_device('cuda')
```

```python
llm=HuggingFacePipeline(pipeline=pipeline, model_kwargs={'temperature':0})
```

```python
model_name = "sentence-transformers/all-mpnet-base-v2"
model_kwargs = {"device": "cuda"}
embeddings = HuggingFaceEmbeddings(model_name=model_name, model_kwargs=model_kwargs)
```

```python
import torch
torch.set_default_device('cuda')
```

```python
chromadb = Chroma.from_documents(documents=text_chunks, embedding=embeddings, persist_directory="chroma_db")
retriever = chromadb.as_retriever()
```

```python
retrieverQA = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",
    retriever=retriever,
    verbose=True
)
```

```python
def run_rag(retrieverQA, text_query):
    print(f"Text Query: {text_query}\n")
    result = retrieverQA.run(text_query)
    print("\nResult: ", result)
    return result
```

```python
file_path = '/content/sample_data/Llama2.csv'
data = pd.read_csv(file_path)
data.head(15)
```

```python
keywords = ["True", "False", "Yes", "No", "yes", "no", "true", "false"]
responses = []
explanation = []
count = 0

for index, row in data.iterrows():
    text_query =row["question"]
    result=run_rag(retrieverQA, text_query )

    output_parser = CommaSeparatedListOutputParser()
    format_instructions = output_parser.get_format_instructions()

    output = output_parser.parse(result)
    parsed_output = output

    # Tüm metni tek bir string olarak birleştir
    full_text = "\n".join(parsed_output)

    # '</s>\n' ifadesinden sonraki kısımda 'Option' ile başlayan cümleyi bul
    separator_index = full_text.find("\n\n")
    if separator_index != -1:
        post_separator_text = full_text[separator_index + len("\n\n"):]  # Ayırıcıdan sonraki metni al
        lines = post_separator_text.split("\n")  # Metni satırlara böl
        for line in lines:
            line = line.strip()  # Her satırı temizle
            if line.startswith("Helpful Answer:") or line.startswith("Helpful Answer: "):
                # İlk 'Option' ile başlayan satırı yazdır ve döngüyü sonlandır
                end_index = line.find(".")
                if end_index != -1:
                    explanation_text = line[:end_index + 1]
                    explanation.append(explanation_text)
                else:
                    explanation_text = line
                    explanation.append(explanation_text)

                # Anahtar kelimeleri kontrol et ve responses'a ekle
                for keyword in keywords:
                    if keyword in explanation_text:
                        responses.append(keyword)
                        break
                break
    else:
        print("Aranan ayırıcı metinde bulunamadı.")

    count += 1
    print(count)

print("İşlem tamamlandı.")
```

```python
import csv

with open('/content/sample_data/Mistral.csv', 'r+', newline='') as csvfile:
    reader = csv.reader(csvfile)  # Okuyucu nesnesi
    all_rows = list(reader)  # Tüm satırları bir listeye dönüştürüyoruz

    if len(all_rows) > 0:
        all_rows[0].append('Mistral_content')
        all_rows[0].append('Mistral')

        # explanations ve responses listesindeki her elemanı ilgili satıra ekliyoruz
        for i in range(max(len(explanation), len(responses))):
            if i + 1 < len(all_rows):
                if i < len(explanation):
                    all_rows[i + 1].append(explanation[i])
                else:
                    all_rows[i + 1].append('')  # Eğer explanations listesi daha kısa ise, boş değer ekliyoruz

                if i < len(responses):
                    all_rows[i + 1].append(responses[i])
                else:
                    all_rows[i + 1].append('')  # Eğer responses listesi daha kısa ise, boş değer ekliyoruz
            else:
                # Eğer explanations veya responses listesi satır sayısından uzunsa, yeni satırlar ekliyoruz
                new_row = [''] * (len(all_rows[0]) - 2)
                if i < len(explanation):
                    new_row.append(explanation[i])
                else:
                    new_row.append('')

                if i < len(responses):
                    new_row.append(responses[i])
                else:
                    new_row.append('')

                all_rows.append(new_row)

        # Dosya imlecini dosyanın başına getiriyoruz
        csvfile.seek(0)

        # Yazar nesnesi oluşturuyoruz
        writer = csv.writer(csvfile)

        # Güncellenmiş satırları dosyaya yazıyoruz
        writer.writerows(all_rows)

        # CSV dosyasının kalan kısmını temizliyoruz (eğer varsa)
        csvfile.truncate()

print("CSV güncelleme tamamlandı.")
```
