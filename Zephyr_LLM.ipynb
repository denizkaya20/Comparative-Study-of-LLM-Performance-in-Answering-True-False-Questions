```python
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline
from langchain.output_parsers import CommaSeparatedListOutputParser
from langchain.prompts import PromptTemplate
from langchain.output_parsers.list import ListOutputParser
```

```python
model = AutoModelForCausalLM.from_pretrained("anakin87/zephyr-7b-alpha-sharded", device_map="auto",load_in_8bit=True)
tokenizer = AutoTokenizer.from_pretrained("anakin87/zephyr-7b-alpha-sharded")
```

```python
pipe = pipeline("text-generation", model=model, tokenizer=tokenizer)
```

```python
responses = []
outputs = []
count = 0

for index, row in data_son.iterrows():
    question_text = data_son.iloc[index]
    if index < len(data_son):
        question_text = data_son.iloc[index]
        messages = [
            {
                "role": "system",
                "content": "Answer the multiple choice questions I will give you seriously.\
                You don't need to explain.\nJust say True or False.\n\
                You can say there are no answers to questions you are unsure of or cannot answer.",
            },
            {"role": "user", "content": f"{question_text['question']}\n"},
        ]

        prompt = pipe.tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
        outputs_from_model = pipe(prompt, max_new_tokens=256, do_sample=True, temperature=0.7, top_k=50, top_p=0.95)
        output_parser = CommaSeparatedListOutputParser()
        format_instructions = output_parser.get_format_instructions()
        output = output_parser.parse(outputs_from_model[0]["generated_text"])
        parsed_output = output
        full_text = "\n".join(parsed_output)
        separator_index = full_text.find("</s>\n")

        if separator_index != -1:
            post_separator_text = full_text[separator_index + len("</s>\n"):]
            lines = post_separator_text.split("\n")
            for line in lines:
                line = line.strip()
                if line.startswith("True") or line.startswith("False"):
                  outputs.append(line)
                  end_index = line.find(".")
                  if end_index != -1:
                        full_sentence = line[:end_index+1].replace("\n", " ")  # Remove newline character
                        full_sentence = line[:end_index+1].replace(".", " ")
                          # Add original generated text to outputs
                        responses.append(line.split()[0])  # Add just "True" or "False" to responses
                  else:
                        full_sentence = line.replace("\n", " ")  # Remove newline character
                        full_sentence = line.replace(".", " ")
                          # Add original generated text to outputs
                        responses.append(line.split()[0])  # Add just "True" or "False" to responses
                  break
            else:
                print("Aranan ayırıcı metinde bulunamadı.")
        count += 1
        print(count)

print("İşlem tamamlandı.")
print("Responses:", responses)
print("Outputs:", outputs)
```

```python
import  csv
with open('/content/sample_data/zephyr_llm.csv', 'r+', newline='') as csvfile:
    reader = csv.reader(csvfile)  # reader nesnesi
    all_rows = list(reader)  # Tüm satırları bir listeye dönüştürüyoruz

    # İlk satıra "zephyr1" ve "zephyr1_content" sütun başlıklarını ekliyoruz
    all_rows[0].append('zephyr1')
    all_rows[0].append('zephyr1_content')

    # responses ve outputs listesindeki her elemanı ilgili satıra ekliyoruz
    for i, (response, output) in enumerate(zip(responses, outputs)):
        if i+1 < len(all_rows):
            all_rows[i+1].append(response)
            all_rows[i+1].append(output)
        else:
            # Eğer responses listesi satır sayısından uzunsa, boş satırlar ekliyoruz
            all_rows.append([''] * (len(all_rows[0])-2) + [response, output])

    # Dosya imlecini dosyanın başına getiriyoruz
    csvfile.seek(0)

    # writer nesnesi oluşturuyoruz
    writer = csv.writer(csvfile)

    # Güncellenmiş satırları dosyaya yazıyoruz
    writer.writerows(all_rows)

    # CSV dosyasının kalan kısmını temizliyoruz (eğer varsa)
    csvfile.truncate()

print("CSV güncelleme tamamlandı.")
```

```python
import pandas as pd
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

true_labels = data['answer']
zephyr_predictions = data['zephyr1']

metrics = {
    'Model': ['Zephyr'],
    'Accuracy': [
        accuracy_score(true_labels, zephyr_predictions)
    ],
    'Precision': [
        precision_score(true_labels, zephyr_predictions)
    ],
    'Recall': [
        recall_score(true_labels, zephyr_predictions)
    ],
    'F1 Score': [
        f1_score(true_labels, zephyr_predictions)
    ]
}

metrics_df = pd.DataFrame(metrics) #Metrikleri daha okunabilir hale getirmek için DataFrame'e çevirdik

print(metrics_df)
```
